```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import cv2
import numpy as np

%matplotlib inline
```

```python
#read in the image
img = ('C:/Users/cnsl/Documents/COMPUTER VISION/DIFFERENT WORK/New folder/BEST (4).jpg')
image = mpimg.imread(img)

plt.imshow(image)
```

```python
#convert image to grayscale for filtering 
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

#display image
plt.imshow(gray, cmap = 'gray')
```

```python
#create the custome kernel 

#edge detection 1
edge1 = np.array([[1,0,-1],[0,0,0],[-1,0,1]])

#edge detection 2
edge2 = np.array([[0,1,0],[1,-4,1],[0,1,0]])

#edge detection 3
edge3 = np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]])

#sharpen 
sharpen = np.array([[1,-1,1],[-1,5,-1],[1,-1,1]])

#box blur
box_blur = np.array([[1,1,1],[1,1,1],[1,1,1]])/9
#alterative mean blur applied direct to the image
box_blur_alt = np.ones((3,3), np.float32)/9


#3 x 3 gaussian blur filter
gblur = np.array([[1,2,1],[2,4,2],[1,2,1]])/16
#alterative gaussian blur applied direct to the image
gblur_alt = cv2.GaussianBlur(gray, (3,3), 0)

#5 x 5 gaussian blur filter
g_blur = np.array([[1,4,6,4,1],[4,16,24,16,4],[6,24,36,24,6],[4,16,24,16,4],[1,4,6,4,1]])/256

#5 x 5 gaussian blur filter
g_blur = np.array([[1,4,6,4,1],[4,16,24,16,4],[6,24,36,24,6],[4,16,24,16,4],[1,4,6,4,1]])/256

#3x3 array forthe edge detection insober y operation
sobel_y = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])

#3x3 array forthe edge detection insober x operation
sobel_x = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
```

```python
#filtered_image_x = cv2.filter2D(gray, -1, sobel_x) 

#filtering the image using filter2D,  which has input: (grayscal image, bit_depth, kernel)
#filtered_image_y = cv2.filter2D(filtered_image_x, -1, sobel_y)
filtered_image_blur = cv2.filter2D(gray, -1, edge1)

filtered_image_x = cv2.filter2D(filtered_image_blur, -1, sobel_x)
filtered_image_y = cv2.filter2D(filtered_image_blur, -1, sobel_y)

#summing the image
final_image1 = cv2.add(filtered_image_x,filtered_image_y)

plt.imshow(final_image1, cmap = 'gray'),plt.title('EDGE DETECTION 1')
```
